```python
# Fourth Exercise: Applying Multiple Filter Criteria to DataFrames
#
# Will be using logical operators: AND (&) OR(|)
```

```python
# Using the logical operator AND (&)
data[(data.State == 'NY') & (data.City == "New York City")].head()
# The Result Set will only show results where
# The City is New York City AND The State is New York
```

```python
# Using the logical operator OR (|)
data[(data.State == 'NJ') | (data.City == "New York City")].head()
# The Result Set will show results where
# The City is New York City OR The State is New Jersey
```

```python
data[(data.State == 'NJ') | (data.City == "New York City") | (data.Time == '20:00')].head()
```

```python
# What is the best way to drop every non-numeric column from a DataFrame?
info = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv")
```

```python
# here we have all the different data type and we want to drop non-numeric columns like object
info.dtypes
```

```python
# As you can see by the result we dropped country which had a data type of object (non-numeric)
import numpy as np
info.select_dtypes(include=[np.number]).dtypes
```

```python
# Clearly from line 89 we only have data with type: object
# Therefore when we try to drop NON-NUMERIC columns we have a
# dataframe as shown by line 90
```
